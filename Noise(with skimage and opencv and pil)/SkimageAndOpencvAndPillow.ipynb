Checking the performance of some python libraries such as skimage,opencv and pillow in Image Processing (with nosiy picture)

```python
import matplotlib.pyplot as plt
from skimage import io
from skimage.filters import gaussian, sobel, hessian, laplace , prewitt , meijering , sato , median
```

```python
image=io.imread("index.jpg")
plt.imshow(image)
```

```python
GaussianFilter=gaussian(image,sigma=2,mode='constant', cval=0.0)
plt.imshow(GaussianFilter)
```

```python
GrayImage=io.imread("index.jpg",as_gray=True)
SobelFilter=sobel(GrayImage)
plt.imshow(SobelFilter, cmap="gray")
```

```python
HessianFilter=hessian(image)
plt.imshow(HessianFilter)
```

```python
LaplacianFilter=laplace(image)
plt.imshow(LaplacianFilter)
```

```python
PrewittFilter=prewitt(image)
plt.imshow(PrewittFilter)
```

```python
SatoFilter=sato(image)
plt.imshow(SatoFilter)
```

```python
MedianFilter=median(image)
plt.imshow(MedianFilter)
```

```python
MeijeringFilter=meijering(image)
plt.imshow(MeijeringFilter)
```

Let's check it with opencv and pillow!

```python
import cv2
import numpy as np
```

```python
Image = cv2.imread("index.jpg",cv2.IMREAD_GRAYSCALE)
 
img_changed = cv2.resize(Image, (600, 500), interpolation= cv2.INTER_LINEAR)
```

```python
plt.imshow(img_changed)
```

First histogram and equalization:

```python
Histogram = cv2.calcHist([Image],[0],None,[256],[0,256])
Equalize = cv2.equalizeHist(Image)
Stacking = np.hstack((Image, Equalize))
```

```python
HI=cv2.equalizeHist(Stacking)
cv2.imwrite('EqualizedOutPut.jpg', Equalize)
```

```python
Result_equ=cv2.calcHist([Equalize],[0],None,[256],[0,256])
```

```python
plt.plot(Histogram)
plt.plot(Result_equ);plt.show()
```

```python
from PIL import Image, ImageFilter
Image = Image.open("index.jpg")
OutPut_1 = Image.filter(ImageFilter.GaussianBlur)
OutPut_2 = Image.filter(ImageFilter.MedianFilter)
OutPut_3 = Image.filter(ImageFilter.BLUR)
OutPut_4 = Image.filter(ImageFilter.SHARPEN)
Stacking_1 = np.hstack((Image ,OutPut_1, OutPut_2))
cv2.imwrite('FirstOutPut.jpg',Stacking_1)
Stacking_2 = np.hstack((Image ,OutPut_3, OutPut_4))
cv2.imwrite('SecondOutPut.jpg',Stacking_2)
```
